## Question 1: Read the crunchbase_odm_orgs.csv CSV file into a Pandas data frame and drop records where company names are None values:
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame


```python
import pandas as pd

df = pd.read_csv("Data/crunchbase_odm_orgs.csv")
shape_before = df.shape

df_clean = df.dropna(subset=['name'])

df_clean.head(5)
```

## Question 2: Create a data frame that contains only the records of USA-based companies whose name starts with "Ac":
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame


```python
df_USA_name_Ac = df_clean[
    (df_clean['country_code'] == 'USA') &
    (df_clean['name'].str.startswith('Ac'))
]

print("record count = ", len(df_USA_name_Ac))

print(df_USA_name_Ac.head(5))
```

## Question 3: Convert the data frame from the previous step into a list of JSON objects:
- Print the count of JSON objects
- Print the first 5 JSON objects
- Write all JSON objects into a text (JSON string) file
- Print the number of records in the resulting file


```python
import json

df_json = df_USA_name_Ac.to_dict(orient='records')

object_count = len(df_json)

print("json object count = ", object_count)

for obj in df_json[:5]:
    print(json.dumps(obj, indent=4))

with open("usa_ac_companies.json", "w") as file:
    json.dump(df_json, file, indent=2)
```

## Question 4: Read the JSON objects from the created file back into a data frame, filter for companies based in New York (city):
- Print the records in the resulting data frame
- Write the output to a webpage using Flask

```python
with open("usa_ac_companies.json", "r") as file:
    data = json.load(file)

df = pd.DataFrame(data)

df_nyc_companies = df[df['city'] == 'New York']

df_nyc_companies
```

```python
from flask import Flask, render_template_string

app = Flask("json app")

@app.route("/")
def show_nyc_companies():
    return df_nyc_companies.to_json(orient='records')

app.run(host='localhost', port=5005)
```

## Question 5: Implement Flask API functions to Return a list of Crunchbase companies based in a user submitted City from the original dataset (4 Points)
- Take a user input through a form to return a list of JSON objects from the city queried

```python
from flask import Flask, request, jsonify, render_template_string
import pandas as pd
import json

df = pd.read_csv("Data/crunchbase_odm_orgs.csv")

# Initialize Flask app
app = Flask("crunchbase")

# Home page with city search form
@app.route("/", methods=["GET", "POST"])
def search_form():
    return """
    <!DOCTYPE html>
    <html>
    <head>
        <title>City Company Search</title>
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css">
    </head>
    <body>
        <div class="container mt-5">
            <h2>Search Crunchbase Companies by City</h2>
            <form action="/results" method="post">
                <div class="mb-3">
                    <label for="city" class="form-label">Enter City Name</label>
                    <input type="text" class="form-control" id="city" name="city" required>
                </div>
                <button type="submit" class="btn btn-primary">Search</button>
            </form>
        </div>
    </body>
    </html>
    """

@app.route("/results", methods=["POST"])
def results():
    city = request.form["city"].strip().lower()

    # Ensure 'city' column exists
    if 'city' not in df.columns:
        return jsonify({"error": "City column not found in data"}), 500

    # Drop rows with missing city names
    df_clean = df.dropna(subset=['city'])

    # Filter for case-insensitive match
    filtered = df_clean[df_clean['city'].str.lower() == city]

    # Convert to list of dicts
    results = filtered.to_dict(orient="records")

    return jsonify(results)


app.run(host='localhost', port=5005)
```
